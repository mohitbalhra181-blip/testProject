```python
class emp:
    def get(self,x,y):
        self.id=x
        self.salary=y
    def show(self):
        print("id=",self.id)
        print("salary=",self.salary)
e1=emp()
e2=emp()
e3=emp()

e1.get(101,10000)
e2.get(202,20000)
e3.get(303,30000)


e1.show()
e2.show()
e3.show()


# e1.id=101
# e2.id=202

# e1.salary=10000
# e2.salary=20000
```
